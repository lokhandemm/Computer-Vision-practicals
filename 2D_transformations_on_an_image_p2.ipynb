```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files

import numpy as np
```

```python
#upload and load image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
img = cv2.cvtColor(cv2.imread(image_path),cv2.COLOR_BGR2RGB)
```

```python
#original image
img = cv2.imread("port.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
#translation
tx, ty = 100,200
translation_matrix = np.float32([[1,0,tx], [0,1, ty]])
img_translation = cv2.warpAffine(img, translation_matrix, (img.shape[1], img.shape[0]))
plt.imshow(img_translation)
plt.axis('off')
plt.show()
```

```python
#rotation by 30
center = (img.shape[1]//2, img.shape[0]//2)
rotation_matrix = cv2.getRotationMatrix2D(center, 30, 1)
rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
plt.imshow(rotated_img)
plt.axis('off')
plt.show()
```

```python
#shearing
img = cv2.imread('port.jpg')
shear_factor = 0.2
shear_matrix = np.array([[1, shear_factor, 0], [shear_factor, 1, 0]], dtype=np.float32)
img_sheared = cv2.warpAffine(img, shear_matrix, (img.shape[1], img.shape[0]))
plt.imshow(img_sheared)
plt.axis('off')
plt.show()
```

```python
#reflection
rows, cols = img.shape[:2]
reflect_matrix = np.float32([[-1, 0, cols], [0, 1, 0]])
reflected = cv2.warpAffine(img, reflect_matrix, (img.shape[1], img.shape[0]))
plt.imshow(reflected)
plt.axis('off')
plt.show()
```
