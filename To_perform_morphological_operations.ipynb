**AIM:** Perform morphological operations
 1.Thresholding 2. Dilation 3.Erosion   4.Opening  5. Closing 6. gradient

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
#import the image
img = cv2.imread('eyeball.jpg', cv2.IMREAD_GRAYSCALE)

_,binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
#define the kernel(jo parts mein sabse jaake multiply hoga)
kernel = np.ones((5, 5), np.uint8)

#dialation
dialation_img = cv2.dilate(binary_img,kernel, iterations=1)
#erosion
erosion_img = cv2.erode(binary_img,kernel, iterations=1)
#opening
opening_img=cv2.morphologyEx(binary_img,cv2.MORPH_OPEN,kernel)
#closing
closing_img=cv2.morphologyEx(binary_img,cv2.MORPH_CLOSE,kernel)
#gradient
gradient_img=cv2.morphologyEx(binary_img,cv2.MORPH_GRADIENT,kernel)
#top hat
tophat_img=cv2.morphologyEx(binary_img,cv2.MORPH_TOPHAT,kernel)
#black hat
blackhat_img=cv2.morphologyEx(binary_img,cv2.MORPH_BLACKHAT,kernel)

title=['original','threshold','dialation','erosion','opening','closing','gradient','tophat','blackhat']
img=[img,binary_img,dialation_img,erosion_img,opening_img,closing_img,gradient_img,tophat_img,blackhat_img]

#mandatory part for all images
plt .figure(figsize=(20,10))
for i in range (9):
  plt.subplot(3,3,i+1)
  plt.imshow(img[i],cmap='gray')
  plt.title(title[i])
  plt.axis('off')
plt.show()

```
